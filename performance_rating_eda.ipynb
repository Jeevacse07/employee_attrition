```python
import pandas as pd

df = pd.read_csv(r"D:\Jeeva\ds_course\employee_attrition\employee_attrition_cleaned.csv")
df.info()
```

```python
df['performancerating'] = df['performancerating'].map(3:1,4:0)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10,15))
sns.countplot(data=df,x="performancerating",hue="jobrole_sales executive")
plt.title("Box plot analysis of performancerating vs jobrole_sales executive")
plt.xlabel("Performance Rating")
plt.xlabel("jobrole_sales executive ")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df1 = df.copy()
df1['yearsincurrentrole_slab'] = pd.cut(df1['yearsincurrentrole'],bins=[0,3,6,9,12,15,18],labels=['0-3','3-6','6-9','9-12','12-15','15-18'])

df_yr = pd.crosstab(
    index=df1["yearsincurrentrole_slab"],
    columns=df1["performancerating"],
    normalize = "index"
)
print(df_yr)

plt.figure(figsize=(10,15))
sns.countplot(data=df1,x="performancerating",hue="yearsincurrentrole_slab")
plt.title("count plot analysis of performancerating  vs yearsincurrentrole")
plt.xlabel("Performance Rating")
plt.ylabel(" yearsincurrentrole ")
plt.grid(True)
plt.show()
```

```python
# Get top 20 features most correlated with 'performancerating' (excluding itself)
top20_features = df.corr()['performancerating'].abs().sort_values(ascending=False).iloc[1:21].index.tolist()
print(top20_features)

df_top20 = df[top20_features]
# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df_top20.corr(),annot=True,cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix: Top 20 Features and performancerating')
plt.show()
```
